```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Загрузка данных

```python
fraud_df = pd.read_parquet("data/transaction_fraud_data.parquet")
```

```python
currency_df = pd.read_parquet("data/historical_currency_exchange.parquet")
```

```python
fraud_df.info()
```

## Предобработка данных

### Преобразуем и разобьем дату

```python
fraud_df["timestamp"] = pd.to_datetime(fraud_df["timestamp"])
fraud_df["date"] = fraud_df["timestamp"].dt.date
fraud_df["hour"] = fraud_df["timestamp"].dt.hour
fraud_df["weekday"] = fraud_df["timestamp"].dt.weekday  # 0=Mon
```

### Развернем last_hour_activity

```python
last = pd.json_normalize(fraud_df["last_hour_activity"])
last.columns = ["lh_"+c for c in last.columns]
fraud_df = pd.concat([fraud_df, last], axis=1)
```

### Нормализуем сумму транзакции

```python
currency_df["date"] = pd.to_datetime(currency_df["date"]).dt.date

# присоединяем курс по дате транзакции
fraud_df = fraud_df.merge(currency_df, left_on="date", right_on="date", how="left", suffixes=("","_cur"))

# функция для нормализации
def to_usd(row):
    cur = row["currency"]
    rate = row.get(cur)
    if pd.isna(rate) or rate == 0:
        return np.nan
    return row["amount"] / rate

fraud_df["amount_usd"] = fraud_df.apply(to_usd, axis=1)
fraud_df["amount_usd"].describe()
```

```python
# Требуемые пакеты:
# pip install lightgbm scikit-learn pandas numpy

import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import average_precision_score, roc_auc_score, classification_report
from sklearn.inspection import permutation_importance
import time

# -------------------------
# Настройки фич (добавлены булевы признаки)
# -------------------------
numeric_features = [
    "hour", "weekday",
    "lh_num_transactions", "lh_total_amount", "lh_unique_merchants",
    "lh_unique_countries", "lh_max_single_amount", "amount_usd",
    # boolean features (включаем как числовые 0/1)
    "is_card_present", "is_outside_home_country", "is_high_risk_vendor", "is_weekend"
]

cat_features = ["country", "city_size", "card_type", "device", "channel"]

# -------------------------
# Препроцессор (numeric -> impute + scale; categorical -> impute + ordinal)
# -------------------------
numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="__missing__")),
    ("ord", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", cat_transformer, cat_features)
    ],
    remainder="drop"  # остальные колонки не нужны
)

# -------------------------
# Pipeline с LGBM (для RandomizedSearch используем n_jobs=1 внутри модели)
# -------------------------
base_lgb = lgb.LGBMClassifier(objective="binary", boosting_type="gbdt",
                              n_jobs=1, random_state=42)

pipe = Pipeline([
    ("pre", preprocessor),
    ("model", base_lgb)
])

# -------------------------
# RandomizedSearchCV параметры (умеренный размер n_iter)
# -------------------------
param_dist = {
    "model__n_estimators": [100, 200, 400],
    "model__learning_rate": [0.01, 0.05, 0.1],
    "model__num_leaves": [31, 63, 127],
    "model__max_depth": [6, 10, 16, -1],
    "model__subsample": [0.6, 0.8, 1.0],
    "model__colsample_bytree": [0.6, 0.8, 1.0],
    "model__reg_alpha": [0.0, 0.1, 1.0],
    "model__reg_lambda": [0.0, 0.1, 1.0]
}

rs = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_dist,
    n_iter=12,                     # уменьшено для скорости
    scoring="average_precision",
    cv=StratifiedKFold(n_splits=3, shuffle=True, random_state=42),
    n_jobs=-1,                     # параллелим по заданиям GridSearch (не по деревьям)
    random_state=42,
    verbose=2
)

# -------------------------
# Разделение данных (time-aware лучше, но здесь используем stratify для простоты)
# -------------------------
# prepared_df должен быть предоставлен пользователем и содержать все перечисленные колонки
# Проверка наличия колонок:
required = set(numeric_features + cat_features + ["is_fraud"])
missing = required.difference(set(prepared_df.columns))
if missing:
    raise ValueError(f"В prepared_df отсутствуют колонки: {missing}")

X = prepared_df.drop(columns=["is_fraud"])
y = prepared_df["is_fraud"].astype(int)

# основной split: train / test
X_train_full, X_test_full, y_train_full, y_test_full = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# Подвыборка для ускорённого тюнинга (тюнинг на 25% train)
X_sub, _, y_sub, _ = train_test_split(
    X_train_full, y_train_full, train_size=0.25, random_state=42, stratify=y_train_full
)

# -------------------------
# Запуск RandomizedSearch на подвыборке
# -------------------------
print("RandomizedSearch на подвыборке (быстро)...")
t0 = time.time()
rs.fit(X_sub, y_sub)
t1 = time.time()
print(f"RandomizedSearch занял {t1-t0:.1f} сек")
print("Лучшие параметры (rs.best_params_):")
print(rs.best_params_)
print("Лучший CV score (AP):", rs.best_score_)

# -------------------------
# Подготовка финального препроцессора и имен фичей
# -------------------------
# Обучаем preprocessor на всем train_full, чтобы получить стабильные имена фич
preprocessor.fit(X_train_full)

# Функция для извлечения имён фичей из ColumnTransformer (упрощённый вариант,
# который корректно обрабатывает наш pipeline: numeric -> same names, ordinal -> same names)
def get_feature_names_from_preprocessor(ct, numeric_cols, cat_cols):
    feature_names = []
    # numeric part -> names = numeric_cols
    # cat part -> OrdinalEncoder -> leaves names as input cat_cols
    # Loop through transformers_ to be safe
    for name, trans, cols in ct.transformers_:
        if name == "remainder":
            continue
        # Если трансформер — pipeline, берём последний шаг для проверки типа
        transformer = trans
        if hasattr(trans, "named_steps"):
            transformer = list(trans.named_steps.values())[-1]
        # OneHotEncoder case (not used here) would require get_feature_names_out
        if hasattr(transformer, "get_feature_names_out") and transformer.__class__.__name__ == "OneHotEncoder":
            # get_feature_names_out expects input_features parameter in newer sklearn
            try:
                names = transformer.get_feature_names_out(cols)
            except Exception:
                # fallback
                names = [f"{c}_{i}" for c in cols]
            feature_names.extend(list(names))
        else:
            # default: keep original column names
            feature_names.extend(list(cols))
    return feature_names

feature_names = get_feature_names_from_preprocessor(preprocessor, numeric_features, cat_features)
print(f"Число итоговых фич: {len(feature_names)}")

# -------------------------
# Трансформируем train/val/test в DataFrame с именами фич
# -------------------------
def transform_to_df(preproc, X_df, feature_names):
    X_trans = preproc.transform(X_df)
    # если sparse -> toarray
    try:
        import scipy.sparse as sp
        if sp.issparse(X_trans):
            X_trans = X_trans.toarray()
    except Exception:
        pass
    return pd.DataFrame(X_trans, columns=feature_names, index=X_df.index)

X_train_trans = transform_to_df(preprocessor, X_train_full, feature_names)
X_test_trans = transform_to_df(preprocessor, X_test_full, feature_names)

# Для early stopping разобьём train на train/val (stratify)
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train_trans, y_train_full, test_size=0.10, random_state=42, stratify=y_train_full
)

# -------------------------
# Финальное обучение LGBM с найденными параметрами (тренируем на DataFrame чтобы избежать предупреждений)
# -------------------------
# Получаем лучшие параметры для LGBM (удаляем префикс 'model__')
best_params = {}
for k, v in rs.best_params_.items():
    if k.startswith("model__"):
        best_params[k.replace("model__", "")] = v

# Устанавливаем разумные fallback-параметры, если чего-то не было в best_params
best_params.setdefault("n_estimators", 400)
best_params.setdefault("learning_rate", 0.05)

final_lgb = lgb.LGBMClassifier(objective="binary", boosting_type="gbdt",
                               n_jobs=-1, random_state=42, **best_params)

```

```python
print("Начинаем финальное обучение LGBM (с early stopping)...")
t0 = time.time()
final_lgb.fit(
    X_tr, y_tr,
    eval_set=[(X_val, y_val)],
    eval_metric="average_precision"
)
t1 = time.time()
print(f"Финальное обучение заняло {t1-t0:.1f} сек")

# -------------------------
# Оценка на тесте
# -------------------------
y_proba = final_lgb.predict_proba(X_test_trans)[:, 1]
y_pred = (y_proba > 0.5).astype(int)

print("Test AP:", average_precision_score(y_test_full, y_proba))
print("Test ROC AUC:", roc_auc_score(y_test_full, y_proba))
print(classification_report(y_test_full, y_pred, digits=4))

# -------------------------
# Permutation importance (на DataFrame с именами фич)
# -------------------------
print("Запускаем permutation importance (параллельно, n_repeats=10)...")
r = permutation_importance(final_lgb, X_test_trans, y_test_full, n_repeats=10, random_state=42, n_jobs=-1)

sorted_idx = r.importances_mean.argsort()[::-1]
print("Top feature importances (permutation):")
for idx in sorted_idx[:30]:
    print(f"{feature_names[idx]}: {r.importances_mean[idx]:.6f} ± {r.importances_std[idx]:.6f}")
```
